**Q8:** Multiple Inheritance ka istemal karte hue do parent classes `Father` aur `Mother` banayein, aur unse aik child class `Child` derive karein jo dono ke habits inherit kare.

```python
class Father:
    def __init__(self, favourite_sport, hobby):
        self.favourite_sports = favourite_sport
        self.hobby = hobby

class Mother:
    def __init__(self, habit):
        self.habit = habit

class Child(Father, Mother):
    def __init__(self, favourite_sport, hobby, habit):
        # Explicitly call each parent constructor
        Father.__init__(self, favourite_sport, hobby)
        Mother.__init__(self, habit)

# Testing the code
c1 = Child("cricket", "Chess", "speaking truth")
print(c1.favourite_sports, c1.hobby, c1.habit)
```

**Q11:** Method Overriding kya hai? Parent class `Bird` banayein jisme `fly()` method ho jo print kare 'Most birds can fly'. Child class `Ostrich` banayein jo is method ko override kare aur print kare 'Ostriches cannot fly'.

```python
class Bird:
    def __init__(self,breed,colour,speed,name):
        self.breed=breed
        self.colour= colour
        self.speed = speed
        self.name= name
    def flying(self):
        print(f"a bird named{self.name} of breed {self.breed} with speed {self.speed} and with colour{self.colour} is flying the sky")
```

```python
class ostrich(Bird):
    def __init__(self,breed,colour,speed,name):
        super().__init__(breed,colour,speed,name)
    def flying(self):
         print(f"a bird named {self.name} of breed {self.breed} with speed {self.speed} and with colour {self.colour} cannot fly in sky")
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
from abc import ABC, abstractmethod
class Payment(ABC):
    def __init__(self, amount):
        self.amount = amount          
    @abstractmethod
    def process_payment(self):
        pass/
        
class CreditCardPayment(Payment):
    def __init__(self, amount, card_number):
        super().__init__(amount)  # Parent class ke amount ko call kiya
        self.card_number = card_number
        
    def process_payment(self):
        print(f"Processing Credit Card payment of Rs.{self.amount} using Card: {self.card_number}")

# 3. Second Subclass: UPIPayment
class UPIPayment(Payment):
    def __init__(self, amount, upi_id):
        super().__init__(amount)  # Parent class ke amount ko call kiya
        self.upi_id = upi_id
        
    def process_payment(self):
        print(f"Processing UPI payment of Rs.{self.amount} via UPI ID: {self.upi_id}")


# --- Testing Both Subclasses (Dono ko test karna zaroori hai) ---

# Credit Card Se Payment Test Karte Hain
payment1 = CreditCardPayment(5000, "1234-5678-9876")
payment1.process_payment()

# UPI Se Payment Test Karte Hain
payment2 = UPIPayment(1200, "user@bank")
payment2.process_payment()

```
