**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   

  
    def get_balance(self):
        return self.__balance

    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited {amount}. New balance: {self.__balance}"
        else:
            return "Deposit amount must be positive"


    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew {amount}. New balance: {self.__balance}"
        else:
            return "Insufficient funds"
```

```python
b1=BankAccount(10000)
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

```python
# Parent class
class Shape:
    def __init__(self, color):
        self.color = color

    def show_color(self):
        print(f"Shape color: {self.color}")

# Child class 1
class Circle(Shape):
    def __init__(self, color, radius):
        # super() se parent ka constructor call
        super().__init__(color)
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

# Child class 2
class Square(Shape):
    def __init__(self, color, side):
        # super() se parent ka constructor call
        super().__init__(color)
        self.side = side

    def area(self):
        return self.side * self.side
```

```python
c1 = Circle("Red", 5)
s1 = Square("Blue", 4)
```

```python
c1.
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
class Number:
    def __init__(self, value):
        self.value = value

 
    def __add__(self, other):
        return Number(self.value + other.value)


n1 = Number(10)
n2 = Number(20)

n3 = n1 + n2   
print(n3.value)

```

**Q15:** Kya Python default roop se Method Overloading (same method name with different arguments) support karta hai? Agar nahi, toh hum same name ke method ko multiple arguments ke sath kaise handle karte hain? Code likhein.

```python
class Calculator:
    def add(self, *args):
 
        if len(args) == 2:
            return args[0] + args[1]
      
        elif len(args) == 3:
            return args[0] + args[1] + args[2]
    
        else:
            return sum(args)

c = Calculator()

print(c.add(10, 20))        
print(c.add(10, 20, 30))    
print(c.add(1, 2, 3, 4, 5)) 

```

**Q16:** `abc` module ka use karte hue aik abstract class `Computer` banayein jisme abstract method `process()` ho. Is class ka object banane ki koshish karein aur dekhein kya hota hai.

```python
from abc import ABC

class Computer(ABC):
    @abstractmethod
    def process(self):
        pass


class Laptop(Computer):
    def process(self):
        print("Laptop is processing...")

class Desktop(Computer):
    def process(self):
        print("Desktop is processing...")


l = Laptop()
l.process()

d = Desktop()
d.process()
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
from abc import ABC, abstractmethod

class Car(ABC):
    def start_engine(self):
        print("Engine has started...")

    
    @abstractmethod
    def drive(self):
        pass

class SportsCar(Car):
    def drive(self):
        print("Driving the sports car fast!")
```

**Q19:** Abstraction aur Encapsulation ke darmiyan kya wazeh farq (main difference) hai? Dono ke concepts ko short statements mei define karein.

```python
from abc import ABC, abstractmethod

#  Abstraction Example
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

    def breathe(self):
        print("All animals breathe...")

class Dog(Animal):
    def sound(self):
        print("Dog barks: Woof Woof!")

class Cat(Animal):
    def sound(self):
        print("Cat meows: Meow Meow!")

# Encapsulation Example
class Student:
    def __init__(self, name, grade):
        self.__name = name       # private variable
        self.__grade = grade     # private variable

    # getter methods
    def get_name(self):
        return self.__name

    def get_grade(self):
        return self.__grade

    # setter method
    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade
            print(f"Grade updated to {self.__grade}")
        else:
            print("Invalid grade!")

```

```python
# Testing Abstraction
dog = Dog()
cat = Cat()

dog.breathe()
dog.sound()

cat.breathe()
cat.sound()
```

```python
# Testing Encapsulation
s1 = Student("Ali", 85)
print("Name:", s1.get_name())
print("Grade:", s1.get_grade())

s1.set_grade(92)   # valid update
s1.set_grade(150)  # invalid update
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.


```python
from abc import ABC, abstractmethod

class Payment(ABC):
    def start_transaction(self):
        print("Transaction started...")  

    @abstractmethod
    def make_payment(self, amount):
        pass   


class CreditCardPayment(Payment):
    def make_payment(self, amount):
        print(f"Paid {amount} using Credit Card.")


class UPIPayment(Payment):
    def make_payment(self, amount):
        print(f"Paid {amount} using UPI.")


```

```python
cc = CreditCardPayment()
cc.start_transaction()
cc.make_payment(50)
```

```python
upi = UPIPayment()
upi.start_transaction()
upi.make_payment(300)
```
